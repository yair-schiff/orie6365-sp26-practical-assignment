### Define Sweep

```python
ns = [10, 100, 1000]
ms = [100, 1000, 10_000]
sigmas = [1e2, 1e4, 1e6]
mus = [1e-2]  # [0, 1e-3, 1e-2]
losses: list[plotting.LossName] = ["quadratic", "logistic"]  # , "l1"]

total_runs = 0
for n in ns:
    for m in ms:
        if n > m:
            continue
        for sigma in sigmas:
            for loss in losses:
                for mu in mus:
                    total_runs += 2
```

### Experiments - Adaptive: `False`

```python
n_iters = 1000
plotting.plot_runs(adaptive=False, ns=ns, ms=ms, sigmas=sigmas, mus=mus, losses=losses, n_iters=n_iters, total_runs=total_runs)
```
